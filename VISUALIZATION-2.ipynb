#### Mean strategy


In this study, the
missing numerical features were replaced by the mean

```python
#One of the technique is mean imputation in which the missing values are replaced with the mean value of the entire feature column
#df.fillna(df.mean())
df['age'] = df['age'].fillna(df['age'].mode()[0])

df['bp'] = df['bp'].fillna(df['bp'].mode()[0])

df['bgr'] = df['bgr'].fillna(df['bgr'].mode()[0])

df['bu'] = df['bu'].fillna(df['bu'].mode()[0])

df['sc'] = df['sc'].fillna(df['sc'].mode()[0])

df['sod'] = df['sod'].fillna(df['sod'].mode()[0])

df['pot'] = df['pot'].fillna(df['pot'].mode()[0])

df['hemo'] = df['hemo'].fillna(df['hemo'].mode()[0])

df['pcv'] = df['pcv'].fillna(df['pcv'].mode()[0])

df['wbcc'] = df['wbcc'].fillna(df['wbcc'].mode()[0])

df['rbcc'] = df['rbcc'].fillna(df['rbcc'].mode()[0])

df



```

Correlation between age and whether a patient has chronic kidney disease

Firstly, let's look at the distribution of affected people depending on their ages 

```python
plt.figure(figsize=(70,25))
plt.legend(loc='upper left')
g = sns.countplot(data = df, x = 'age', hue = 'class')
g.legend(title = 'Class', loc='center left', bbox_to_anchor=(0.1, 0.5), ncol=1)
g.tick_params(labelsize=20)
plt.setp(g.get_legend().get_texts(), fontsize='32')
plt.setp(g.get_legend().get_title(), fontsize='42')
g.axes.set_title('Graph of age vs number of patients with chronic kidney disease',fontsize=50)
g.set_xlabel('age',fontsize=40)
g.set_ylabel("count",fontsize=40)
```

let we see the corrolation matrix between the age and the number of affected cases 

```python
import matplotlib.pyplot as plt 
import seaborn as sns
sns.countplot(df['class'],label="Count")
plt.show()
```

```python
age_corr = ['age', 'class']
age_corr1 = df[age_corr]
age_corr_y = age_corr1[age_corr1['class'] == 1].groupby(['age']).size().reset_index(name = 'count')
age_corr_y.corr()
```

```python
sns.regplot(data = age_corr_y, x = 'age', y = 'count').set_title("Correlation graph for Age vs chronic kidney disease patient")
```

```python
age_corr_n = age_corr1[age_corr1['class'] == 0].groupby(['age']).size().reset_index(name = 'count')
age_corr_n.corr()
```

```python
sns.regplot(data = age_corr_n, x = 'age', y = 'count').set_title("Correlation graph for Age vs healthy patient")
```

conclusion:
    There is a weak positive correlation between age and chronic kidney disease patients. We obtained an R value of approximately 0.387 and an R-square value of approximately 0.150. This means that only 15% of variation can be explained by the relationship between the 2 variables.
